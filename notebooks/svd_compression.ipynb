```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np 
```

```python
import cv2

image = cv2.imread('cat_ofcourse.png', cv2.IMREAD_GRAYSCALE)

plt.figure()
plt.imshow(image, cmap='gray')
plt.title('mandatory cat pic')
plt.show()
```

```python
U, S, Vt = np.linalg.svd(image, full_matrices=False)
U.shape, S.shape, Vt.shape
```

```python
reconstructed_image = np.matrix(U[:,:1]) * np.diag(S[:1]) * np.matrix(Vt[:1,:])
plt.figure()
plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image, n=1')
plt.show()
```

```python
plt.figure(figsize=(8,3))

start, end, step = 5, 40, 10
for i in range(start, end, step):
    plt.subplot(1, (end - start) // step + 1, (i - start) // step + 1)
    reconstructed = np.matrix(U[:, :i]) * np.diag(S[:i]) * np.matrix(Vt[:i, :])
    plt.imshow(reconstructed, cmap='gray')
    plt.title('n = %s' % i)

#plt.tight_layout()
plt.show()
```
